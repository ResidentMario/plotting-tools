`seaborn` is a convenient, flexible, and terrifically well-thought out data exploration -targeted data visualization library.

Let's see a few highlights.

First, a bar chart.

The before:

```python
df = collisions['BOROUGH'].value_counts()
df.plot.bar(figsize=(12, 8))
```

The after:

```python
import seaborn as sns
```

```python
sns.countplot(collisions['BOROUGH'])
```

`countplot` does the counting aggregation for you. That is a really useful shortcut.

Plus, check out what the API reference has to say about data formats:

> Input data can be passed in a variety of formats, including:
> * Vectors of data represented as lists, numpy arrays, or pandas Series objects passed directly to the x, y, and/or hue parameters.
> * A “long-form” DataFrame, in which case the x, y, and hue variables will determine how the data are plotted.
> * A “wide-form” DataFrame, such that each numeric column will be plotted.
> * Anything accepted by plt.boxplot (e.g. a 2d array or list of vectors)

In other words, no crazy `melt`-`pivot`-`unstack` shenanigans. This really lowers the barrier to entry of plotting your data and makes things a lot easier.

(as an aside, `seaborn` isn't very consistent about providing resizing options, so we'll have to reach into another library called `matplotlib` to do that. more on that later!)

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 8))
sns.countplot(collisions['BOROUGH'])
```

Another thing that's important to be aware of: `seaborn` injects its own style into the notebook. This modifies other plotters which also use `matplotlib`. If we redraw the `pandas` chart, here's what we get:

```python
df = collisions['BOROUGH'].value_counts()
df.plot.bar(figsize=(12, 8))
```

Usually the redraw is quite a bit nicer-looking, but YMMV.

```python
df = (collisions
          .groupby(['CONTRIBUTING FACTOR VEHICLE 1', 'NUMBER OF PERSONS INJURED'])
          .count()
          .loc[:, 'DATE']
          .rename('COUNT(COLLISIONS)')
          .to_frame()
          .unstack('NUMBER OF PERSONS INJURED'))

df.columns = df.columns.droplevel(0)

df = (df.rename_axis("INJURIES", axis='columns')
         .rename_axis("CONTRIBUTING FACTOR", axis='rows')
         .apply(lambda srs: srs.set_value("2+", srs.iloc[2:].sum()), axis=1)
         .apply(lambda srs: srs.drop(srs.iloc[2:-1].index.values), axis=1)
         .fillna(0)
         .drop("Unspecified")
         .sort_values(by=0, ascending=False)
         .head(5))
```

```python
df.plot.bar(figsize=(12,8))
```

```python
top_five_collision_factors = collisions['CONTRIBUTING FACTOR VEHICLE 1'].value_counts().iloc[1:6].index

collisions_of_interest = collisions[
    (collisions['CONTRIBUTING FACTOR VEHICLE 1'].isin(top_five_collision_factors)) &
    (collisions['NUMBER OF PERSONS INJURED'] <= 2)
]

collisions_of_interest.head()
```

Note that all we've had to do is *select* the records we're interesting in. We didn't have to *transform* the data at all.

```python
plt.figure(figsize=(12, 8))

sns.barplot(x='CONTRIBUTING FACTOR VEHICLE 1', 
            y='NUMBER OF PERSONS INJURED',
            hue='NUMBER OF PERSONS INJURED',
            estimator=len,
            data=collisions_of_interest)
```

This is much cleaner and much more easily understandable. We didn't have to invoke any hard-to-do data transformations, we just selected what we needed using boolean indexing.

Let's keep going with the comparisons. `seaborn` doesn't have a histogram (well, it does, but not as a separate plot *per se*) but does have something cooler, a KDEplot (which, to be fair, `pandas` also has):

```python
df = (collisions.groupby('ZIP CODE')
        .sum()
        .loc[:, 'NUMBER OF PERSONS INJURED']
        .rename('COUNT(COLLISIONS)'))
```

Re-using the `df` here from earlier because it wasn't too bad...

```python
plt.figure(figsize=(12, 8))

sns.kdeplot(data=df)
```

Let's look at the `seaborn` box plot. Before:

```python
df = collisions[['CONTRIBUTING FACTOR VEHICLE 1', 'NUMBER OF PERSONS INJURED']]
df = df[df['CONTRIBUTING FACTOR VEHICLE 1'] != 'Unspecified']

subsrs = []
df.groupby('CONTRIBUTING FACTOR VEHICLE 1')\
    .apply(lambda df: subsrs.append(pd.Series(df['NUMBER OF PERSONS INJURED'].values, name=df.name)))
    
df = pd.concat(subsrs, axis='columns')
```

```python
df.iloc[:, :5].plot.box(figsize=(12, 8))
```

After. Note, this time instead of picking the first five names we'll just reuse the list of top five collision factors we generated earlier.

```python
plt.figure(figsize=(12, 8))

collisions_of_interest = collisions[
    (collisions['CONTRIBUTING FACTOR VEHICLE 1'].isin(top_five_collision_factors))
]

sns.boxplot(x='CONTRIBUTING FACTOR VEHICLE 1', y='NUMBER OF PERSONS INJURED', data=collisions_of_interest)
```

This data is admittedly not very "boxplot-y", because it's a discrete and not continuous numerical range.

Nevertheless, very quickly, here's a `violinplot`, a boxplot improvement using KDEs which is `seaborn`-specific.

```python
plt.figure(figsize=(12, 8))

collisions_of_interest = collisions[
    (collisions['CONTRIBUTING FACTOR VEHICLE 1'].isin(top_five_collision_factors))
]

sns.violinplot(x='CONTRIBUTING FACTOR VEHICLE 1', y='NUMBER OF PERSONS INJURED', data=collisions_of_interest)
```

`seaborn` brings an ease of use to these kinds of exploratory chart types which is not present in `pandas`. This means that in general:

If you have a quick and easy plotting need, you can use `pandas` built-in plotting.

For more complicated ideas, `seaborn` is what you need to use.

However, since these are both what I will from now on call "high-level plotting libraries", using either/or effectively demands that you know what features they provide reasonably well. So if you're not familiar with these two tools, I encourage you to, on your own, go visit their documentation, browse through what you find, and take a day or whatever to apply them to your favorite dataset(s).
