Next, let's look at something that actually goes about and implements `bokeh` (and does a bunch of other things besides): `holoviews`. `holoviews` is a high-level plotting library designed for evil geniuses which takes all of the innovations in the current Python plotting stack and turns them all up to 11:



```python
from ggplot import meat
```

```python
import pandas as pd
import numpy as np
```

```python
import holoviews as hv
hv.notebook_extension()
```

```python
meat['date'] = pd.to_datetime(meat['date'])
```

```python
meat.head()
```

```python
df = meat['beef']
```

```python
hv.Scatter(df)
```

```python
hv.Scatter(df) + hv.Curve(df)
```

```python
hv.Scatter(df) * hv.Curve(df)
```

```python
%opts Curve plot[aspect=2, bgcolor=None]

hv.Scatter(df)
```

```python
%opts Curve plot[aspect=2]

hv.Scatter(df)(style={'color': 'steelblue'}) * hv.Curve(df)(style={'color': 'lightsteelblue'})
```

```python
%opts Layout [fig_size=300] Scatter [aspect=2] Curve style(linewidth=1)

hv.Scatter(df)(style={'color': 'steelblue'}) * hv.Curve(df)(style={'color': 'lightsteelblue'})
```

```python
%opts Layout [fig_size=300] Scatter [aspect=2]

hv.Scatter(df)(style={'color': 'steelblue'})
```
